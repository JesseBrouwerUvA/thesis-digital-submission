```python
# Find the ffmpeg path on your device
ffmpeg_exe = Path(imageio_ffmpeg.get_ffmpeg_exe())

# Find the folder in which the ffmpeg program is
ffmpeg_folder = ffmpeg_exe.parent

# Add that folder to system path so it can be used
os.environ["PATH"] = str(ffmpeg_folder) + os.pathsep + os.environ["PATH"]

print("ffmpeg location:", ffmpeg_exe)
```
